```python
# Importaci贸n de librer铆as
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from imblearn.pipeline import Pipeline
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder, label_binarize
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc, roc_auc_score

# Configuraciones
pd.set_option('display.max_columns', None)
sns.set(style='whitegrid')
```


##  Carga de Datos

Cargamos el archivo `locales_procesado.csv` con las variables ya preprocesadas.


```python

# Carga del archivo CSV
data = pd.read_csv('../data/locales_procesado.csv')
data.head()
```


##  Exploraci贸n de Datos

Visualizamos algunas estad铆sticas generales y la distribuci贸n de la variable objetivo.


```python

# Estad铆sticas generales
data.describe()

# Distribuci贸n de la variable objetivo
data['valoracion'].value_counts()
```


##  Transformaci贸n de la Variable Objetivo

Dado que `valoracion` es continua, la transformamos en categor铆as para clasificaci贸n.


```python

# Crear categor铆as a partir de la valoraci贸n
bins = [0, 3, 4, 5]
labels = ['baja', 'media', 'alta']
data['valoracion_categoria'] = pd.cut(data['valoracion'], bins=bins, labels=labels, include_lowest=True)

# Revisar la distribuci贸n
print(data['valoracion_categoria'].value_counts())
```


## 锔 Divisi贸n de Datos

Dividimos el conjunto en entrenamiento y prueba de forma estratificada para mantener la proporci贸n de clases.


```python

# Variables predictoras y objetivo
X = data.drop(columns=['valoracion', 'valoracion_categoria'])
y = data['valoracion_categoria']

# Divisi贸n de datos
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
```


##  Preprocesamiento y Pipeline

Aplicamos escalado a las variables num茅ricas y one-hot encoding a las categ贸ricas.


```python
# Columnas
numeric_features = ['latitud', 'longitud', 'valoracion_norm', 'density_1000m', 
                    'ratio_500m_2km', 'dist_city_center_km', 
                    'puntuacion_std', 'reviews_std']
categorical_features = ['categoria_negocio']

# Preprocesador
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features),
        ('cat', OneHotEncoder(drop='first', sparse_output=False), categorical_features)
    ]
)

# Pipeline con SMOTE integrado
pipeline = Pipeline(steps=[
    ('prep', preprocessor),
    ('smote', SMOTE(random_state=42)),
    ('rf', RandomForestClassifier(n_estimators=200, max_depth=6, random_state=42))
])
```


##  Entrenamiento del Modelo

Entrenamos el modelo Random Forest con el conjunto balanceado.


```python
pipeline.fit(X_train, y_train)
```

##  Evaluaci贸n del Modelo

Evaluamos el rendimiento del modelo mediante un reporte de clasificaci贸n y la matriz de confusi贸n.

```python
# Predicciones
y_pred = pipeline.predict(X_test)

# Reporte de Clasificaci贸n
print("Classification Report:\n")
print(classification_report(y_test, y_pred))

# Matriz de Confusi贸n
plt.figure(figsize=(6,5))
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel("Predicci贸n")
plt.ylabel("Real")
plt.title("Matriz de Confusi贸n")
plt.savefig("../images/matriz_confusion.png", dpi=300, bbox_inches='tight')
plt.show()
```

##  Curvas ROC y AUC

Visualizamos las curvas ROC y calculamos el AUC para cada clase en un esquema One-vs-Rest.

```python
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_curve, auc
import numpy as np

# Binarizar las clases
y_test_bin = label_binarize(y_test, classes=np.unique(y_test))
y_score = pipeline.predict_proba(X_test)

n_classes = y_test_bin.shape[1]
fpr = dict()
tpr = dict()
roc_auc = dict()

plt.figure(figsize=(8,6))
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])
    plt.plot(fpr[i], tpr[i], lw=2, label=f'Clase {i} (AUC = {roc_auc[i]:.2f})')

plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlim([-0.02, 1.0])
plt.ylim([0.0, 1.02])
plt.xlabel('Tasa de Falsos Positivos')
plt.ylabel('Tasa de Verdaderos Positivos')
plt.title('Curvas ROC - One-vs-Rest')
plt.legend(loc="lower right")
plt.savefig("../images/curvas_roc.png", dpi=300, bbox_inches='tight')
plt.show()
```

###  Interpretaci贸n de la Curva ROC y AUC

La Curva ROC permite evaluar la capacidad del modelo para distinguir entre clases en un escenario One-vs-Rest.  
- **AUC cercano a 1.0** indica un buen desempe帽o.  
- **AUC menor a 0.5** puede indicar que el modelo predice peor que un clasificador aleatorio.  
En nuestro caso, debemos analizar la clase "alta" con atenci贸n, ya que su AUC es bajo.

 **Conclusi贸n:**  
Si la AUC es baja, es recomendable:
- Revisar el balance de clases.  
- A帽adir nuevas variables que mejoren la discriminaci贸n.  
- Ajustar los hiperpar谩metros del modelo.

##  Conclusiones

A partir de las m茅tricas obtenidas, el modelo presenta:
- Buen rendimiento en las clases mayoritarias.
- Posibles 谩reas de mejora en las clases minoritarias.
- AUC promedio aceptable en el conjunto de test.

Esto nos ayudar谩 a entender c贸mo ajustar el modelo en futuras iteraciones.

##  Importancia de Variables

Analizamos la importancia relativa de las variables utilizadas por el modelo Random Forest para comprender su contribuci贸n a las predicciones.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Extraer el modelo y el preprocesador del pipeline
rf_model = pipeline.named_steps['rf']
preprocessor = pipeline.named_steps['prep']

# Obtener nombres de las variables tras preprocesamiento
features = preprocessor.get_feature_names_out()

# Importancia de variables
importances = rf_model.feature_importances_
feature_names = pipeline.named_steps['prep'].get_feature_names_out()

feat_importance_df = pd.DataFrame({
    'Feature': feature_names,
    'Importance': importances
}).sort_values(by='Importance', ascending=False)

plt.figure(figsize=(8, 5))
plt.barh(feat_importance_df['Feature'], feat_importance_df['Importance'])
plt.xlabel('Importancia')
plt.ylabel('Variable')
plt.title('Importancia de Variables - Random Forest')
plt.gca().invert_yaxis()  # Para que la m谩s importante quede arriba
plt.tight_layout()
plt.savefig("../images/importancia_variables.png", dpi=300, bbox_inches='tight')
plt.show()
```

锔 Importante:

Esto nos permite:

```python
# Visualizaci贸n de la distribuci贸n de clases antes y despu茅s de SMOTE

# Distribuci贸n de clases original
plt.figure()
sns.countplot(x=y_train, palette='pastel')
plt.title("Distribuci贸n de Clases Original")
plt.xlabel("Clase")
plt.ylabel("Cantidad")
plt.savefig("../images/distribucion_clases_original.png", dpi=300, bbox_inches='tight')
plt.show()

# Preprocesar los datos para convertirlos a num茅rico
X_train_transformed = preprocessor.fit_transform(X_train)

# Aplicar SMOTE a los datos transformados (solo para visualizaci贸n)
smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train_transformed, y_train)

# Distribuci贸n de clases despu茅s de SMOTE
plt.figure()
sns.countplot(x=y_train_res, palette='muted')
plt.title("Distribuci贸n de Clases Despu茅s de SMOTE (Visualizaci贸n)")
plt.xlabel("Clase")
plt.ylabel("Cantidad")
plt.savefig("../images/distribucion_clases_smote.png", dpi=300, bbox_inches='tight')
plt.show()
```
